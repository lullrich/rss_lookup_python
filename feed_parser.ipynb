```python
import feedparser
import pandas as pd
```

```python
def get_feed_info(url):
    feed_info = {}
    print("getting info for %s" % (url))
    try:
        feed = feedparser.parse(url)["feed"]
        feed_info["url"] = url
        feed_info["link"] = feed.get('link', None)
        feed_info["title"] = feed.get('title', None)
        feed_info["subtitle"] = feed.get('subtitle', None)
        feed_info["language"] = feed.get('language', None)
        feed_info["description"] = feed.get('description', None)
        feed_info["category"] = feed.get('category', None)
        feed_info["updated"] = feed.get('updated', None)
    except Exception as e:
        print(e)
    return feed_info
        
```

```python
urls = pd.read_csv("rss_links.csv", header=None, names=["urls"])
len(urls)
```

```python
rss_infos = (get_feed_info(url) for url in urls["urls"])
```

```python
%time df = pd.DataFrame.from_dict(rss_infos)
```

```python
df.to_csv("test.csv", sep="\t", encoding="UTF-8")
```
